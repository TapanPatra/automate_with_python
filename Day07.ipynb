### 1. Create your daily checklist

Let's start by defining a list of tasks you want to accomplish for the day.

```python
daily_checklist = [
    "Finish the project report",
    "Send follow-up emails",
    "Attend team meeting",
    "Review design mockups",
    "Plan next week's schedule",
    "Exercise for 30 minutes"
]

print("Your daily checklist:")
for i, task in enumerate(daily_checklist):
    print(f"{i+1}. {task}")
```

### 2. Mark tasks as completed or incomplete

Now, let's go through your checklist. For each task, indicate if you completed it (by typing 'y' or 'yes') or not (by typing 'n' or 'no').

```python
completed_tasks = []
incomplete_tasks = []

print("\n--- Task Review ---")
for task in daily_checklist:
    while True:
        response = input(f"Did you complete '{task}'? (y/n): ").strip().lower()
        if response in ['y', 'yes']:
            completed_tasks.append(task)
            break
        elif response in ['n', 'no']:
            incomplete_tasks.append(task)
            break
        else:
            print("Invalid input. Please enter 'y' for yes or 'n' for no.")
```

### 3. Overview of your day

Here's a summary of your completed and incomplete tasks for the day.

```python
print("\n--- Daily Summary ---")

print("\nCompleted Tasks:")
if completed_tasks:
    for i, task in enumerate(completed_tasks):
        print(f"{i+1}. {task}")
else:
    print("No tasks completed today.")

print("\nIncomplete Tasks:")
if incomplete_tasks:
    for i, task in enumerate(incomplete_tasks):
        print(f"{i+1}. {task}")
else:
    print("All tasks completed today! Great job!")
```
