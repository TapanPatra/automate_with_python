# **Default Params**

```python
def hello(name = "Guest"):
  print(f"Hello, {name}!")
```

```python
hello()
```

```python
hello("Tapan")
```

# **Global Variable**

```python
x=100
def print_x():
  print(x)
```

```python
print_x()
```

```python
y=100
def print_y():
  y=10
  print(y)
```

```python
print_y()
```

```python
y=100
def print_y():
  global y
  y=20
  print(y)
```

```python
print_y()
```

```python
print(y)
```
